```python
# Question 4
def isPalindrome(s:str):
    reverse = ''.join(reversed(s))
    if s == reverse:
        return True
    return False

isPalindrome('zooz')
```

```python
# Question 5
def combine(a:list, b:list):
  result = []
  for i in range(len(a)):
    result.append(a[i])
    result.append(b[i])
  return result

combine(['a','b','c'],[1,2,3])
```

```python
# Question 6
def Fibonacci():
  Fib_list = [1,1]
  for i in range(2,100):
    temp = Fib_list[i-2] + Fib_list[i-1]
    Fib_list.append(temp)
  for i in range(0,100):
    print(i, Fib_list[i])
  return Fib_list

Fibonacci()
```

```python
# Question 7
def leapyear(year:int):
  if year % 100 != 0:
    if year % 4 == 0:
      return True
    else:
      return False
  elif year % 400 == 0:
    return True
  else:
    return False

leapyear(1900)
```
